```python


import os
import shutil

import google.generativeai as genai

# Setup Gemini API (Replace with your actual API key)
genai.configure(api_key="API_KEY_HERE")

# Create the model
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}

# Instantiate the Generative Model with specific configurations
model = genai.GenerativeModel(
  model_name="gemini-1.5-flash",  # Choose the model you want to use
  generation_config=generation_config,
)

# Define directory paths
data_folder = "name_of_the_folder_containing_code_files"
commented_data_folder = "name_of_the_folder_to_save_commented_code_files"

# Initialize counter for processed files
count = 0

# Ensure the output folder exists
if not os.path.exists(commented_data_folder):
    os.makedirs(commented_data_folder)


# Traverse through the directory to process each file
for root, dirs, files in os.walk(data_folder):
  for filename in files:
    if filename.endswith((".html", ".ts", ".kt")):
        count += 1
        print("Working on file ", filename , " count: ", count)
        
        # Construct the full file path
        file_path = os.path.join(root, filename)
        # Read the code file
        with open(file_path, "r") as file:
            code_content = file.read()

        # Save the commented code to the new folder, with the same folder structure
        commented_file_path = os.path.join(commented_data_folder, file_path[len(data_folder)+1:])
        
        # Ensure the output folder exists
        if not os.path.exists(os.path.dirname(commented_file_path)):
            os.makedirs(os.path.dirname(commented_file_path))

        # if the file does not exist, create it else move to the next file
        if not os.path.exists(commented_file_path):
            print(f"Commenting and saving: {filename}")
        else:
            print(f"File already exists: {filename}")
            continue

        # Construct the prompt for adding necessary comments in the code file
        prompt = f"Do not change the existing code or comments and add 
                   necessary and important in line comments to this code
                   file so that it is easier for a user to understand.        
                   \n\n{code_content}"

        chat_session = model.start_chat(
          history=[]
        )
        # Generate comments with Gemini API
        response = chat_session.send_message(prompt)
        commented_code = response.text

        # remove the first line and the last line
        commented_code = commented_code.split("\n")[1:-1]
        commented_code = "\n".join(commented_code)

        # Save the commented code to the new folder
        with open(commented_file_path, "w") as file:
          file.write(commented_code)
       

print("Comment addition process completed!")
print(f"Total files commented: {count}")
```
